# **Text Pre-Processing:**

```python
import nltk
import re
import string
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, ne_chunk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
def Text_Preprocessing(text):
  # to lowecase:
  text = text.lower()

  # remove numbers:
  text = re.sub(r'\d+', '', text)

  # remove the punctuation:
  translator = str.maketrans('', '', string.punctuation)
  text = text.translate(translator)

  # tokenize:
  tokens = word_tokenize(text)

  # remove stiowords:
  stop_words = set(stopwords.words('english'))
  tokens = [word for word in tokens if word not in stop_words]

  # stemming or lammatization:
  # porter = PorterStemmer()
  # new_text = [porter.stem(word) for word in tokens]
  lemma = WordNetLemmatizer()
  new_text = [lemma.lemmatize(word, pos='v') for word in tokens]
  new_text = ' '.join(new_text)

  return new_text
```

```python
text = """Hello All, Good Morning! Hi I am Dibyendu Biswas, I an 23 years old. 
        I am from Siliguri, West bengal. my emial id id: dibyendubiswas1998@gmail.com
        I am playing football and eating you!!!! Bye@!!!"""
        
Text_Preprocessing(text)
```
