```python
import nltk
import re
import string
```

# **Using nltk:**

# **1. Convert To Lower Case:**

```python
def TO_Lowercase(text):
  return text.lower()

TO_Lowercase("Hello, I Am Dibyendu Biswas, From Siliguri, West Bengal")
```

# **2. Remove Numbers:**

```python
def remove_Numbers(text):
  txt = re.sub(r'\d+', '', text)
  print(txt)

remove_Numbers("hello, I am Dibyendu, I am 23 years old")
remove_Numbers("My Phone Numbers is: 9785696556")
```

# **3. Convert Numbers to Text:**

```python
def convert_numbers(text):
  import inflect
  q = inflect.engine()

  temp_str = text.split()
  new_str = []

  for word in temp_str:
    if word.isdigit():
      temp = q.number_to_words(word)
      new_str.append(temp)
    else:
      new_str.append(word)

  final_str = ' '.join(new_str)
  print(final_str)


convert_numbers("Hello, I am Dibyendu Biswas, I am 23 years old")
convert_numbers("he has 1000 dogs")
convert_numbers("I am born in year of 1999 12 march")
convert_numbers(" my date of birth is: 1999-03-12")
convert_numbers(" my date of birth is: 1999 03 12")
```

# **4. Remove Punctuation:**

```python
def remove_punctuation(text):
  translator = str.maketrans('', '', string.punctuation)
  txt = text.translate(translator)
  print(txt)

remove_punctuation("Hello, Good Morning!, It's Okay Buy!, Hmm.")
```

# **5. Word Tokenization:**

```python
def word_tokenization(text):
  from nltk.tokenize import word_tokenize
  nltk.download('punkt')
  
  tokens = word_tokenize(text)
  print(tokens)

word_tokenization("Hey, I am Dibyendu Biswas, I am from siliguri, West Bengal. Today I am start to learn NLP")
```

```python
def custome_word_tokenization(text):
  tokens = text.split(' ')
  print(tokens)

custome_word_tokenization("Hey, I am Dibyendu Biswas, I am from siliguri, West Bengal. Today I am start to learn NLP")
```

# **6. Sentence Tokenization:**

```python

def sentence_tokenization(text):
  from nltk.tokenize import sent_tokenize
  nltk.download('punkt')

  sentence_tokens = sent_tokenize(text)
  print(sentence_tokens)

sentence_tokenization("Narrative paragraphs are the building blocks you need to tell a story in English. This type of paragraph explains something that happened. Think about a time that a friend or family member came to you and said, “guess what!”. They probably then told you a story about something that happened to them.")
```

# **7. Remove Stopwords(default):**

```python
def remove_stopwords(text):
  from nltk.corpus import stopwords
  from nltk.tokenize import word_tokenize
  nltk.download('stopwords')
  nltk.download('punkt')

  stopwords = set(stopwords.words('english'))
  tokens = word_tokenize(text)
  filtered_text = [word for word in tokens if word not in stopwords]
  print(filtered_text)

remove_stopwords("This is Dibyendu Biswas. I am from Siliguri")
```

# **8. Stemming:**

PotterStemmer, LandCasteStemmer, SnowballStemmer

## **i. Stemming For Word:**

```python
def stemming_for_words(word):
  from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
  porter = PorterStemmer()
  lancaste = LancasterStemmer()
  snowball = SnowballStemmer('english')

  print(f"Using Porter Stemmer {word}: {porter.stem(word)}")
  print(f"Using Lancaste Stemmer {word}: {lancaste.stem(word)}")
  print(f"Using Snowball Stemmer {word}: {snowball.stem(word)} \n")

stemming_for_words('runs')
stemming_for_words('running')
stemming_for_words('run')

stemming_for_words('Run')
stemming_for_words('Runs')
stemming_for_words('Running')
```

## **ii. Stemming For Sentences:**

```python
def stemming_for_stentence(text):
  from nltk.tokenize import word_tokenize
  from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
  nltk.download('punkt')

  porter = PorterStemmer()
  lancaste = LancasterStemmer()
  snowball = SnowballStemmer('english')
  tokens = word_tokenize(text)

  porter_stem = [porter.stem(word) for word in tokens]
  lancaste_stem = [lancaste.stem(word) for word in tokens]
  snowball_stem = [snowball.stem(word) for word in tokens]

  print(porter_stem)


stemming_for_stentence("I am going to school. I am now travelling")
```

# **9. Lemmatization:**

## **i. Word Lemmatization:**

```python
def word_lemmatization(word, verb):
  from nltk.stem import WordNetLemmatizer
  nltk.download('wordnet')
  lemma = WordNetLemmatizer()
  new_word = lemma.lemmatize(word, pos=verb)
  print(new_word)

word_lemmatization('runs', 'v')
word_lemmatization('running', 'v')

word_lemmatization('runs', 'n')
word_lemmatization('running', 'a')
```

## **ii. Sentence Lemmatization:**

```python
def sentence_lemmatization(text, pos):
  from nltk.tokenize import word_tokenize
  from nltk.stem import WordNetLemmatizer
  nltk.download('wordnet')
  lemma = WordNetLemmatizer()
  tokens = word_tokenize(text)

  lemmatize_word = [lemma.lemmatize(word, pos) for word in tokens]
  print(lemmatize_word)

sentence_lemmatization("I am going to school. I am travelling now!", 'v')
sentence_lemmatization("I am going to school. I am travelling now!", 'a')
sentence_lemmatization("I am going to school. I am travelling now!", 'n')
```

# **10. Parts of Speech (POS):**

```python
def Pos_Taggs(text):
  from nltk.tokenize import word_tokenize
  from nltk import pos_tag
  nltk.download('averaged_perceptron_tagger')

  tokens = word_tokenize(text)
  pos_tags = pos_tag(tokens)
  print(pos_tags)

Pos_Taggs("I am Dibyendu, from siliguri, West Bengal")
```

# **11. Name Entity Recognization (NER):**

```python
def Name_Entity_Recognization(text):
  import nltk
  from nltk.tokenize import word_tokenize
  from nltk import pos_tag, ne_chunk
  nltk.download('maxent_ne_chunker')
  nltk.download('words')

  tokens = word_tokenize(text)
  word_tag = pos_tag(tokens)
  NER = ne_chunk(word_tag)
  print(NER)

Name_Entity_Recognization("I am Dibyendu, from Siliguri, Darjeeling, West Bengal")
```

# **12. Word Count:**

```python
def Word_Count1(txt):
    import nltk
    from nltk.tokenize.regexp import WhitespaceTokenizer # white space

    tokens = WhitespaceTokenizer().tokenize(txt)
    print(f"All Tokens: {tokens}")
    print(f"Length of tokens:  {len(tokens)}")

    # mow find the distinct tokens (vocabulary) and it's length:
    my_vocab = set(tokens)
    print(f"My Vocabulary is: {my_vocab}")
    print(f"Length of Vocabulary:  {len(my_vocab)}")

Word_Count1("There is no need to panic. We need to work together, take small yet important measures to ensure self-protection,” the Prime Minister tweeted.")
```

# **13. Frequency Distribution:**

```python
def Frequency_Distributions(txt):
    import nltk
    from nltk.tokenize.regexp import WordPunctTokenizer
    from nltk import FreqDist

    tokens = WordPunctTokenizer().tokenize(txt)

    fre = FreqDist(tokens)
    dict = {key:vals for key, vals in zip(fre.keys(), fre.values())}
    print(dict)

Frequency_Distributions("There is no need to panic. We need to work together, take small yet important measures to ensure self-protection,” the Prime Minister tweeted.")
```

```python
def Plot_Frequency_Distribution(text):
  import matplotlib.pyplot as plt
  import nltk
  from nltk.tokenize.regexp import WordPunctTokenizer
  from nltk import FreqDist

  tokens = WordPunctTokenizer().tokenize(text)
  freqDist = FreqDist(tokens)
  
  plt.figure(figsize=(15,6))
  freqDist.plot()
  plt.show()


Plot_Frequency_Distribution("There is no self need to self panic. We need to work self together, take small yet self important measures to self ensure self-protection,” the self Prime Minister tweeted self.")
```
